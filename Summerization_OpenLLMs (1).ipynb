##### The articles scrapped from different news sites. Here, I scrapped from CNN and manorama(Kerala's media) site.

```python
import requests
from bs4 import BeautifulSoup
```

```python
url1 = "https://edition.cnn.com/2023/10/27/asia/south-korea-teachers-strike-analysis-intl-hnk/index.html"

response = requests.get(url1)

# Check if the request was successful
if response.status_code == 200:

    soup = BeautifulSoup(response.text, 'html.parser')

    article_head1 = ""

    # Find the headline of the article
    headline = soup.find("h1", class_="headline__text")
    if headline:
        article_head1 += headline.get_text() + "\n\n"

    # Find and extract paragraphs
    paragraphs = soup.find_all("p", class_="paragraph")
    article1 =""
    for paragraph in paragraphs:
        article1 += paragraph.get_text() + "\n\n"
    article1 = article_head1 + article1
    print(article1)
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
url1 = "https://www.onmanorama.com/travel/travel-news/2023/10/27/drivers-unions-launch-spot-taxi-call-centre.html"

response = requests.get(url1)

# Check if the request was successful
if response.status_code == 200:

    soup = BeautifulSoup(response.text, 'html.parser')

    article_head2 = ""
    article2 = ""

    # Find the headline of the article
    headline = soup.find("h1", class_="story-headline")
    if headline:
        article_head2 += headline.get_text() + "\n\n"

    # Find and extract paragraphs
    paragraphs = soup.find_all("div", class_="story-body article-content")
    for paragraph in paragraphs:
        article2 += paragraph.get_text() + "\n\n"
        article2 = article_head2 + article2

    print(article2)
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

# Modeling and Fine Tuning

```python
# Three models are imported
from transformers import T5ForConditionalGeneration, T5Tokenizer
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from transformers import BertTokenizerFast, EncoderDecoderModel

T5_PATH = 'google/flan-t5-base'
# Initialize the T5 model and tokenizer
t5_model = T5ForConditionalGeneration.from_pretrained(T5_PATH)
t5_tokenizer = T5Tokenizer.from_pretrained(T5_PATH)

GPT_PATH = 'gavin124/gpt2-finetuned-cnn-summarization-v1'
# Initialize the GPT-2 model and tokenizer
gpt2_model = GPT2LMHeadModel.from_pretrained(GPT_PATH)
gpt2_tokenizer = GPT2Tokenizer.from_pretrained(GPT_PATH)
BERT_PATH = 'mrm8488/bert-small2bert-small-finetuned-cnn_daily_mail-summarization'
# Initialize the BERT model and tokenizer
bert_model = EncoderDecoderModel.from_pretrained(BERT_PATH)
bert_tokenizer = BertTokenizerFast.from_pretrained(BERT_PATH)
```

```python
# Summary with each models
for index, row in data.iterrows():
    article_title = row['Title']
    article_summary = row['Summary']

    # Generate a summary using the T5 model
    input_text = f"summarize: {article_title}\n\n{article_summary}"
    input_ids = t5_tokenizer.encode(input_text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = t5_model.generate(input_ids, max_length=150, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)
    t5_summary = t5_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # Generate a summary using the GPT-2 model
    input_text = f"{article_title}\n\n{article_summary}"
    input_ids = gpt2_tokenizer.encode(input_text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = gpt2_model.generate(input_ids, max_length=150, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)
    gpt2_summary = gpt2_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # Generate a summary using the BERT model
    input_text = f"{article_title}\n\n{article_summary}"
    input_ids = bert_tokenizer.encode(input_text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = bert_model.generate(input_ids, max_length=150, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)
    bert_summary = bert_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # Print or store the generated summaries
    print("Title:", article_title)
    print("T5 Summary:", t5_summary)
    print("GPT-2 Summary:", gpt2_summary)
    print("BERT Summary:", bert_summary)
    print("\n")
```

```python
from rouge import Rouge

rouge = Rouge()

t5_scores = []

for index, row in data.iterrows():
    reference_summary = row['Summary']  # summary
    article_text = row['Title']  # Article content

    # Generate a T5 summary for the article
    input_text = f"summarize: {article_text}"
    input_ids = t5_tokenizer.encode(input_text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = t5_model.generate(input_ids, max_length=150, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)
    t5_summary = t5_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # Convert t5_summary and reference_summary to strings
    t5_summary = t5_summary if isinstance(t5_summary, str) else ' '.join(t5_summary)
    reference_summary = reference_summary if isinstance(reference_summary, str) else ' '.join(reference_summary)

    t5_score = rouge.get_scores(t5_summary, reference_summary, avg=True)
    t5_scores.append(t5_score)


# Print or store the ROUGE scores for the T5 model
for i, score in enumerate(t5_scores):
    print(f"Article {i + 1} ROUGE Scores for T5:")
    print("ROUGE-1:", score["rouge-1"])
    print("ROUGE-2:", score["rouge-2"])
    print("ROUGE-L:", score["rouge-l"])
    print("\n")
```

```python
gpt2_scores = []

for index, row in data.iterrows():
    reference_summary = row['Summary']  # Reference summary
    article_text = row['Title']  # Article content

    # Generate a GPT-2 summary for the article
    input_text = f"{article_text}"
    input_ids = gpt2_tokenizer.encode(input_text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = gpt2_model.generate(input_ids, max_length=150, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)
    gpt2_summary = gpt2_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # Convert gpt2_summary and reference_summary to strings
    gpt2_summary = gpt2_summary if isinstance(gpt2_summary, str) else ' '.join(gpt2_summary)
    reference_summary = reference_summary if isinstance(reference_summary, str) else ' '.join(reference_summary)

    gpt2_score = rouge.get_scores(gpt2_summary, reference_summary, avg=True)
    gpt2_scores.append(gpt2_score)

# Print or store the ROUGE scores for the GPT-2 model
for i, score in enumerate(gpt2_scores):
    print(f"Article {i + 1} ROUGE Scores for GPT-2:")
    print("ROUGE-1:", score["rouge-1"])
    print("ROUGE-2:", score["rouge-2"])
    print("ROUGE-L:", score["rouge-l"])
    print("\n")
```

```python
from summarizer import Summarizer

sentence =['traffic police have been instructed to slap heavy fines for illegal parking','traffic restrictions imposed on heavy goods vehicles on wayanad ghat road']
bert_scores = []

for index, row in data.iterrows():
    reference_summary = row['Summary']  # summary
    article_text = row['Title']  # Article content

    # Summarize the article using the BERT-based summarizer
    bert_summary = Summarizer(article_text)

    # Convert bert_summary and reference_summary to strings
    bert_summary = ' '.join(sentence)
    reference_summary = reference_summary if isinstance(reference_summary, str) else ' '.join(reference_summary)

    bert_score = rouge.get_scores(bert_summary, reference_summary, avg=True)
    bert_scores.append(bert_score)

# Print or store the ROUGE scores for the BERT model
for i, score in enumerate(bert_scores):
    print(f"Article {i + 1} ROUGE Scores for BERT:")
    print("ROUGE-1:", score["rouge-1"])
    print("ROUGE-2:", score["rouge-2"])
    print("ROUGE-L:", score["rouge-l"])
    print("\n")
```
